```python
import numpy as np 
import pandas as pd 
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from boruta import BorutaPy
```

```python
train_data = pd.read_csv(r"Data\train.csv")
test_data = pd.read_csv(r"Data\test.csv")
```

```python
train_data = train_data.drop('Unnamed: 0',axis = 1)
test_data = test_data.drop('Unnamed: 0',axis = 1)
```

```python
train_data['train-test'] = "train"
test_data['train-test'] = "test"
```

```python
combined_df = pd.concat([train_data,test_data]).fillna('TBC')
```

```python
combined_df.to_csv('./Data/CombinedData.csv')
```

```python
scaler = MinMaxScaler()

combined_df["Feature 1"] = scaler.fit_transform(np.array(combined_df["Feature 1"]).reshape(-1,1))
combined_df["Feature 22"] = scaler.fit_transform(np.array(combined_df["Feature 22"]).reshape(-1,1))
combined_df["Feature 23"] = scaler.fit_transform(np.array(combined_df["Feature 23"]).reshape(-1,1))
combined_df["Feature  24"] = scaler.fit_transform(np.array(combined_df["Feature  24"]).reshape(-1,1))
combined_df["Feature 25"] = scaler.fit_transform(np.array(combined_df["Feature 25"]).reshape(-1,1))
combined_df["Feature 26"] = scaler.fit_transform(np.array(combined_df["Feature 26"]).reshape(-1,1))
```

```python
combined_df.T
```

```python
df_group = combined_df.groupby(['train-test'])
train_data = df_group.get_group('train')
test_data = df_group.get_group('test')
```

```python
train_data = train_data.drop('train-test',axis=1)
test_data = test_data.drop('train-test',axis=1)
```

```python
test_data
```

```python
target_variable = train_data['price']
train_data_input = train_data.drop('price',axis = 1)
```

```python
#target_variable = np.array(target_variable, dtype = int)
#train_data_input = np.array(train_data_input,dtype= float)
```

```python
rf = RandomForestRegressor(n_estimators=100, max_depth=5, random_state=42)

# let's initialize Boruta
feat_selector = BorutaPy(
    verbose=2,
    estimator=rf,
    n_estimators='auto',
    max_iter=26  # number of iterations to perform
)

# train Boruta
# N.B.: X and y must be numpy arrays
feat_selector.fit(train_data_input.values.astype(float),target_variable.values.astype(int))

# print support and ranking for each feature
print("\n------Support and Ranking for each feature------")
for i in range(len(feat_selector.support_)):
    if feat_selector.support_[i]:
        print("Passes the test: ", train_data_input.columns[i],
              " - Ranking: ", feat_selector.ranking_[i])
    else:
        print("Doesn't pass the test: ",
              train_data_input.columns[i], " - Ranking: ", feat_selector.ranking_[i])
```

```python
### initialize hits counter
hits = np.zeros((len(train_data_input.columns)))
### repeat 20 times
for iter_ in range(30):
   ### make X_shadow by randomly permuting each column of X
   np.random.seed(iter_)
   X_shadow = train_data_input.apply(np.random.permutation)
   X_boruta = pd.concat([train_data_input, X_shadow], axis = 1)
   ### fit a random forest (suggested max_depth between 3 and 7)
   rf = RandomForestRegressor(max_depth = 5, random_state = 42)
   rf.fit(X_boruta, target_variable)
   ### store feature importance
   feat_imp_X = rf.feature_importances_[:len(train_data_input.columns)]
   feat_imp_shadow = rf.feature_importances_[len(train_data_input.columns):]
   ### compute hits for this trial and add to counter
   hits += (feat_imp_X > feat_imp_shadow.max())
```

```python
col = train_data_input.columns
a = len(train_data_input.columns)
for i in range(2,a):
    if (hits[i] == 0):
        print(col[i])
        train_data_input = train_data_input.drop(col[i],axis = 1)
```

```python
train_data_input = train_data_input.drop('Feature 2',axis = 1)
train_data_input = train_data_input.drop('Feature 3',axis = 1)
train_data_input = train_data_input.drop('Feature 4',axis = 1)
train_data_input = train_data_input.drop('Feature 5',axis = 1)
train_data_input = train_data_input.drop('Feature 6',axis = 1)
train_data_input = train_data_input.drop('Feature 11',axis = 1)
train_data_input = train_data_input.drop('Feature 15',axis = 1)
train_data_input = train_data_input.drop('Feature 18',axis = 1)
train_data_input = train_data_input.drop('Feature 19',axis = 1)
train_data_input = train_data_input.drop('Feature 20',axis = 1)
train_data_input = train_data_input.drop('Feature 21',axis = 1)
#train_data_input = train_data_input.drop('Feature 23',axis = 1)
```
